```python
import requests
import json

url = "https://api.hrrcn.io/v3.4/classification"

payload = json.dumps({
  "clientSKU": "XQEE897656R",
  "clientDescription": "Keep It Classy beauty & skincare",
  "sourceCountry": "US",
  "sourceLanguage": "en"
})
headers = {
  'Content-Type': 'application/json',
  'X-Api-Key': '5udz8EZyjW2mB6kqN8HSv1AwlcckoCiKZDbdhNoh'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
```

```python
import pandas as pd

raw = pd.read_json(response.text)
```

```python
from pandas.io.json import json_normalize

json_struct = json.loads(raw['classifications'].to_json(orient="records"))
df_flat = pd.io.json.json_normalize(json_struct) #use pd.io.json
df_flat[:5]
```
